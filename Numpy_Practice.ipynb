```python
# chapter 3 - Indexing and slicing

# indexing - single element
# slicing - multiple element

# fancy indexing - using list of index
# boolean masking - giving condition and then filtering

import numpy as np

"""
array [index] #1d array
array [row, column] #2d array

"""
arr = np.array([10,20,30,40,50])

print(arr[0])
print(arr[1])
print(arr[3])
print(arr[-1])





```

```python
"""
slicing - start,stop-excluded,step-default-1
arr[start:stop:step]
arr[start:end], start to end -1
negative step, -1 reverse

"""
import numpy as np

arr = np.array([10,20,30,40,50,60])

print(arr[1:5])  # start to end -1
print(arr[:4])  # begining to end -1
print(arr[::2])  # pickes every second element
print(arr[::-1])  # reverse





```

```python
# fancy indexing - selecting multiple elements at once

import numpy as np

arr = np.array([10,20,30,40,50,60])

print(arr[[0,2,4]])  # its a copy of original list, here original list is not changed

```

```python
# boolean masking - filtering data using conditions

import numpy as np

arr = np.array([10,20,30,40,50,60])

print(arr[arr > 25])
```

```python
# Reshaping and manipulating array
# 1. reshaping - modifying dimension without modifying the data, like 1d to 2d or 3d. It can change rows and columns

"""
reshape(rows, columns) specify new shape
if dimensions match

"""

import numpy as np

arr = np.array([1,2,3,4,5,6])
reshaped_arr = arr.reshape(2,3)  # does not create copy of original list, it returns a view
print(reshaped_arr)

```

```python
# 2. flatten 
"""
.ravel() - return views
.flatten() - return copy

"""

import numpy as np

arr_2d = np.array([[1,2,3],[4,5,6]])
print(arr_2d.ravel())
print(arr_2d.flatten())
```

```python
# chapter 5 - broadcasting and vectorization

"""
broadcasting - expand small arrays to larger arrays, so they can match
faster than loops
use - 1d to 2d broadcasting

"""

prices = [100,200,300]
discount = 10   # 10% discount

final_prices = []

for price in prices:
    final_price = price - (price * discount/100)
    final_prices.append(final_price)

print(final_prices)

```

```python
# loops in slow in large datasets
# broadcasting - without loop

import numpy as np

prices = np.array([100,200,300])
discount = 10

final_prices = prices - (prices * discount/100)
print(final_prices)
```

```python
# how numpy handle arrays of diff shapes
# rule 1 : matching dimension, eg:  [1,2,3] + [4,5,6] = [5,7,9]
# rule 2 : expanding single elements, eg: [1,2,3] + 10 = [11,12,13]
# rule 3 : incompatible shapes, eg: [1,2,3] + [5,6] = error

import numpy as np

arr = np.array([100,200,300])
result = arr * 2
print(result)
```

```python
# chapter 6 - Handling Missing & Special Values

"""
like to solve divisible by zero error handling
have 3 built-in functions
1. np.isnan - detect missing values, nan - not a number
2. np.nan_to_num() 
3. np.isinf() - detect infinite values

"""
# 1. isnan - detect missing values

import numpy as np

arr = np.array([1,2,np.nan,4,np.nan,6])

print(np.isnan(arr))

"""
can we compare nan values ? = NO, we cannot compare it directly

"""

```

```python
# 2. nan_to_num, default - 0

import numpy as np

arr = np.array([1,2,np.nan,4,np.nan,6])

cleaned_arr = np.nan_to_num(arr, nan = 100)  # replace nan with 100, default - 0
print(cleaned_arr)
 
```

```python
# 3. infinite

# np.isinf(array) , eg: 10 ** 1000 or 1/0

import numpy as np

arr = np.array([1,2,np.inf, 4, -np.inf, 6])

print(np.isinf(arr))



```

```python
# replace
# np.isinf(array) , eg: 10 ** 1000 or 1/0

import numpy as np

arr = np.array([1,2,np.inf, 4, -np.inf, 6])

print(np.isinf(arr))

cleaned_arr = np.nan_to_num(arr, posinf = 1000, neginf = -1000)
print(cleaned_arr)
```
